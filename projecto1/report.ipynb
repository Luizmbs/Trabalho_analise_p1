```python
import matplotlib.pyplot as plt

# tamanho das instncias
ns = [10, 20, 30, 50, 100, 200, 400, 600, 800, 1000]

# tempos medidos (em segundos)
time_values = [0.0444277, 0.0260961, 0.0439964, 0.0414469, 0.0480979, 0.0521984, 0.0767413, 0.212904, 0.420164, 0.84052]

# complexidade terica O(n^3)
f_values = [n**3 for n in ns]

plt.figure(figsize=(9,5))

# Plot da linha
plt.plot(f_values, time_values, marker='o', linewidth=2)

# ===========================
# Funo simples para evitar sobreposio
# ===========================
label_positions = []   # guarda os y usados para evitar colises
base_offset = 0.03     # quanto subir o rtulo acima do ponto
extra_offset = 0.03    # offset extra quando existir sobreposio


for x, y, z in zip(f_values, time_values, ns):

    # posio inicial sugerida
    label_y = y + base_offset

    # verificao de coliso com outras labels
    while any(abs(label_y - other_y) < 0.03 for other_y in label_positions):
        label_y += extra_offset  # sobe mais at no colidir

    label_positions.append(label_y)

    # desenha o rtulo j ajustado
    plt.text(
        x, label_y,
        f"{z}: {y:.3f}s",
        fontsize=9,
        ha='center',
        va='bottom'
    )

# eixos e ttulo
plt.xlabel("f(n, m)")
plt.ylabel("Time (s)")
plt.title("Tempo de execuo vs complexidade terica")
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
